Este notebook demuestra el uso del m√≥dulo `utils.py`, que convierte cadenas de texto en im√°genes PNG  
usando **matrices binarias** como representaci√≥n interna de cada car√°cter.

## üîç ¬øQu√© hace este proyecto?

El m√≥dulo `utils.py` implementa un sistema de **tipograf√≠a de p√≠xeles** desde cero, sin fuentes externas.  
El flujo completo es:

```
texto  ‚Üí  normalizar  ‚Üí  matrices de letras  ‚Üí  concatenar  ‚Üí  escalar  ‚Üí  colorear  ‚Üí  PNG
```

Cada paso del notebook ilustra una parte de ese pipeline usando las funciones definidas en `utils.py`.

---
## Paso 1 ‚Äî Importar el m√≥dulo y explorar el alfabeto

Antes de trabajar con im√°genes hay que importar todas las funciones del m√≥dulo `utils.py`.  
Tambi√©n se importan `numpy`, `matplotlib` y `PIL` para visualizar resultados dentro del notebook.

### Funciones y objetos usados en este paso

| Nombre | Tipo | Descripci√≥n |
|---|---|---|
| `ALFABETO` | `dict` | Diccionario que mapea cada car√°cter a su matriz binaria 7√óN |
| `normalizar_texto` | funci√≥n | Convierte el texto a caracteres soportados |
| `texto_a_matriz` | funci√≥n | Concatena las matrices de letras en una sola imagen |
| `escalar_matriz` | funci√≥n | Amplia la imagen √óN con nearest-neighbor |
| `matriz_a_imagen` | funci√≥n | Convierte la matriz binaria a imagen PIL RGB |
| `guardar_imagen` | funci√≥n | Pipeline completo: texto ‚Üí PNG |

````python
# ‚îÄ‚îÄ Importar funciones del m√≥dulo propio ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from utils import (
    ALFABETO,           # diccionario {caracter: ndarray}
    normalizar_texto,   # texto ‚Üí lista de chars soportados
    texto_a_matriz,     # texto ‚Üí matriz binaria 2-D
    escalar_matriz,     # ampliar imagen √óN (nearest-neighbor)
    matriz_a_imagen,    # matriz binaria ‚Üí imagen PIL RGB
    guardar_imagen,     # pipeline completo texto ‚Üí PNG
    _color_a_rgb,       # validador de formato de color (demo de errores)
)

# Mostrar cu√°ntos caracteres hay en el alfabeto
print(f"‚úÖ M√≥dulo importado correctamente.")
print(f"   Caracteres soportados: {len(ALFABETO)}")
print(f"   Lista: {sorted(ALFABETO.keys())}")
````

---
## Paso 2 ‚Äî Visualizar una sola letra como matriz e imagen

Aqu√≠ se accede directamente al `ALFABETO` para ver c√≥mo se representa una letra internamente:  
- Cada fila de la matriz corresponde a una fila de p√≠xeles (alto fijo = 7).  
- Cada columna corresponde a un p√≠xel de ancho (variable por letra).  
- `0` = fondo (p√≠xel apagado), `1` = letra (p√≠xel encendido).

Luego se escala la matriz con `escalar_matriz` para que sea visible, y se convierte en imagen  
con `matriz_a_imagen`.

### Funciones usadas

| Funci√≥n | Qu√© hace aqu√≠ |
|---|---|
| `ALFABETO[letra]` | Obtiene la matriz binaria cruda de la letra |
| `escalar_matriz(mat, 10)` | Convierte cada p√≠xel en un bloque 10√ó10 (nearest-neighbor) |
| `matriz_a_imagen(mat, cl, cf)` | Asigna colores a la matriz y devuelve una imagen PIL |

````python
letra = "H"
letra = letra.lower()
# Obtener la matriz binaria de la letra desde el diccionario
mat = ALFABETO[letra]
print(f"Matriz de '{letra}'  ‚Üí  {mat.shape[0]} filas √ó {mat.shape[1]} columnas:")
print(mat)

fig, axes = plt.subplots(1, 2, figsize=(7, 3))

# ‚îÄ‚îÄ Panel izquierdo: matriz num√©rica ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# imshow con cmap='gray_r' muestra 0=blanco, 1=negro
axes[0].imshow(mat, cmap="gray_r", vmin=0, vmax=1)
axes[0].set_title(f"Matriz binaria  '{letra}'  ({mat.shape[0]}√ó{mat.shape[1]})")
# Mostrar la cuadr√≠cula de p√≠xeles con sus coordenadas
axes[0].set_xticks(range(mat.shape[1]))
axes[0].set_yticks(range(mat.shape[0]))
axes[0].grid(True, color="lightblue", linewidth=0.5)

# ‚îÄ‚îÄ Panel derecho: imagen ampliada ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# escalar_matriz usa np.kron para replicar cada p√≠xel en un bloque 10√ó10
mat_grande = escalar_matriz(mat, factor=10)
# matriz_a_imagen asigna colores: 1‚Üícolor_letra, 0‚Üícolor_fondo
img_letra = matriz_a_imagen(mat_grande, color_letra="black", color_fondo="white")
axes[1].imshow(img_letra)
axes[1].set_title(f"Imagen ampliada  '{letra}'  √ó10")
axes[1].axis("off")

plt.tight_layout()
plt.show()
````

---
## Paso 3 ‚Äî Generar la palabra `"hola"` y guardarla

Se usa `guardar_imagen`, la funci√≥n de m√°s alto nivel del m√≥dulo.  
Internamente llama en orden a:

1. `normalizar_texto("hola")` ‚Üí `['h','o','l','a']`  
2. `texto_a_matriz(...)` ‚Üí matriz binaria con letras concatenadas y m√°rgenes  
3. `escalar_matriz(..., factor=10)` ‚Üí ampliar √ó10  
4. `matriz_a_imagen(...)` ‚Üí aplicar colores  
5. `Image.save(ruta)` ‚Üí guardar PNG en disco

### Par√°metros configurables de `guardar_imagen`

| Par√°metro | Valor usado | Efecto |
|---|---|---|
| `color_letra` | `"blue"` | P√≠xeles encendidos en azul (nombre matplotlib) |
| `color_fondo` | `"yellow"` | Fondo amarillo |
| `escala` | `10` | Cada p√≠xel original ‚Üí bloque 10√ó10 |
| `margen_*` | `2` | 2 px de margen en cada borde (antes de escalar) |

````python
# guardar_imagen = pipeline completo en una sola llamada
# Crea la carpeta 'outputs/' autom√°ticamente si no existe
guardar_imagen(
    "hola",                          # texto a renderizar
    "outputs/hola_blue_yellow.png",  # ruta de salida
    color_letra="blue",              # color de p√≠xeles encendidos (nombre)
    color_fondo="yellow",            # color de fondo (nombre)
    escala=10,                       # ampliar √ó10
    margen_sup=2, margen_inf=2,      # m√°rgenes verticales (en px antes de escalar)
    margen_izq=2, margen_der=2,      # m√°rgenes horizontales
)

# Cargar y mostrar el PNG guardado
img = Image.open("outputs/hola_blue_yellow.png")
print(f"Tama√±o de la imagen: {img.size[0]}√ó{img.size[1]} px")
plt.figure(figsize=(8, 2))
plt.imshow(img)
plt.title("'hola' ‚Äî azul sobre amarillo  (nombre matplotlib)")
plt.axis("off")
plt.show()
````

---
## Paso 4 ‚Äî Texto m√°s complejo: `"UNAL 2025"`

Se prueba con may√∫sculas, n√∫meros y un espacio.  
`normalizar_texto` maneja cada car√°cter individualmente:
- Letras may√∫sculas ‚Üí busca directamente en `ALFABETO` (claves A‚ÄìZ definidas).  
- D√≠gitos `0`‚Äì`9` ‚Üí tambi√©n est√°n en `ALFABETO`.  
- Espacio `" "` ‚Üí se inserta un separador m√°s ancho (`espacio_palabras`).

### Par√°metros nuevos usados

| Par√°metro | Valor | Efecto |
|---|---|---|
| `color_letra` | `(0,0,0)` | Negro en formato tupla RGB |
| `color_fondo` | `(255,255,255)` | Blanco en formato tupla RGB |
| `espacio_palabras` | `4` | 4 px de separaci√≥n entre palabras |
| `margen_*` | `3` | Margen m√°s grande que el ejemplo anterior |

````python
# normalizar_texto convierte cada car√°cter; si no existe en ALFABETO ‚Üí '?'
texto_comp= 'UNAL 2025'
texto_comp = texto_comp.lower()
chars = normalizar_texto(texto_comp)
print(f"Texto normalizado: {chars}")

# texto_a_matriz concatena las matrices de cada car√°cter
# con el espaciado y m√°rgenes indicados
guardar_imagen(
    "UNAL 2025",
    "outputs/unal2025.png",
    color_letra=(0, 0, 0),           # negro ‚Äî formato tupla RGB (R,G,B)
    color_fondo=(255, 255, 255),     # blanco ‚Äî formato tupla RGB
    espacio_letras=1,                # 1 px entre letras consecutivas
    espacio_palabras=4,              # 4 px extra al encontrar un espacio
    escala=10,
    margen_sup=3, margen_inf=3,
    margen_izq=3, margen_der=3,
)

img = Image.open("outputs/unal2025.png")
print(f"Tama√±o: {img.size[0]}√ó{img.size[1]} px")
plt.figure(figsize=(12, 2))
plt.imshow(img)
plt.title("'UNAL 2025' ‚Äî negro sobre blanco (RGB)")
plt.axis("off")
plt.show()
````

---
## Paso 5 ‚Äî Efecto de diferentes m√°rgenes (escala √ó10)

Se llama a `texto_a_matriz` directamente (sin guardar) para construir la matriz con distintos  
valores de margen, luego se escala y colorea para mostrar el efecto visual.

`texto_a_matriz` a√±ade filas/columnas de ceros al borde de la imagen **antes** de escalar,  
por lo que el margen final en la imagen guardada es `margen √ó factor` p√≠xeles.

### Funciones usadas en detalle

| Funci√≥n | Par√°metro clave | Efecto |
|---|---|---|
| `texto_a_matriz` | `margen_sup/inf/izq/der` | A√±ade fondo alrededor del texto |
| `escalar_matriz` | `factor=10` | Ampl√≠a toda la matriz, incluidos los m√°rgenes |
| `matriz_a_imagen` | `color_letra`, `color_fondo` | Colorea p√≠xeles 1 y 0 respectivamente |

````python
margenes = [0, 2, 5]   # valores a comparar
fig, axes = plt.subplots(1, len(margenes), figsize=(14, 3))

for ax, m in zip(axes, margenes):
    # texto_a_matriz devuelve la matriz binaria 2-D con m√°rgenes incluidos
    mat = texto_a_matriz(
        "hola",
        margen_sup=m, margen_inf=m,
        margen_izq=m, margen_der=m,
    )
    print(f"Margen={m}  ‚Üí  matriz {mat.shape[0]}√ó{mat.shape[1]}")

    # escalar_matriz: np.kron replica cada p√≠xel en bloque factor√ófactor
    mat_esc = escalar_matriz(mat, factor=10)

    # matriz_a_imagen: ndarray binario ‚Üí imagen PIL RGB con colores aplicados
    img = matriz_a_imagen(mat_esc, color_letra="navy", color_fondo="lightyellow")
    ax.imshow(img)
    ax.set_title(f"Margen = {m} px  ‚Üí  imagen {img.size[0]}√ó{img.size[1]}")
    ax.axis("off")

plt.suptitle("Comparaci√≥n de m√°rgenes ‚Äî escala √ó10", fontsize=13)
plt.tight_layout()
plt.show()
````

---
## Paso 6 ‚Äî Tres combinaciones de color

`guardar_imagen` (y por debajo `_color_a_rgb`) acepta colores en **tres formatos**:

| Formato | Ejemplo | C√≥mo se procesa |
|---|---|---|
| **Nombre** | `"blue"`, `"yellow"` | `matplotlib.colors.to_rgba(nombre)` |
| **Tupla RGB** | `(0,0,0)`, `(255,255,255)` | Validaci√≥n directa de 3 enteros 0‚Äì255 |
| **Hex** | `"#1E90FF"`, `"#F5F5F5"` | `matplotlib.colors.to_rgba(hex)` |

Internamente todos se convierten a `(R, G, B)` antes de construir la imagen.

````python
# Definir las tres configuraciones: (texto, ruta, color_letra, color_fondo, t√≠tulo, formato)
configuraciones = [
    ("hola", "outputs/c1_nombres.png",  "blue",      "yellow",        "1 ‚Äî Nombres"),
    ("hola", "outputs/c2_rgb.png",      (0,0,0),     (255,255,255),   "2 ‚Äî Tupla RGB"),
    ("hola", "outputs/c3_hex.png",      "#1E90FF",   "#F5F5F5",       "3 ‚Äî Hex #RRGGBB"),
]

fig, axes = plt.subplots(1, 3, figsize=(14, 3))

for ax, (texto, ruta, cl, cf, titulo) in zip(axes, configuraciones):
    # guardar_imagen valida el formato de cl y cf con _color_a_rgb antes de pintar
    guardar_imagen(
        texto, ruta,
        color_letra=cl,
        color_fondo=cf,
        escala=10,
        margen_sup=2, margen_inf=2,
        margen_izq=2, margen_der=2,
    )
    img = Image.open(ruta)
    ax.imshow(img)
    ax.set_title(titulo)
    ax.axis("off")

plt.suptitle("Tres formatos de color soportados por guardar_imagen()", fontsize=12)
plt.tight_layout()
plt.show()
````

---
## Paso 7 ‚Äî Validaci√≥n de colores inv√°lidos

`_color_a_rgb` es la funci√≥n interna que valida y convierte cada color.  
Lanza `ValueError` con un mensaje descriptivo si el formato es incorrecto.

Esto evita que un error de tipeo en el color llegue hasta PIL y genere un error  
dif√≠cil de interpretar.

### Casos que genera error

| Entrada | Motivo del error |
|---|---|
| `(300, 0, 0)` | Componente RGB fuera del rango 0‚Äì255 |
| `"colornulo"` | Nombre de color no reconocido por matplotlib |
| `"#ZZZZZZ"` | Hex con caracteres inv√°lidos |
| `42` | Tipo entero, no str ni tupla |

````python
# _color_a_rgb es la funci√≥n de bajo nivel que usa guardar_imagen internamente
# Se import√≥ al inicio para mostrar los mensajes de error directamente

casos_invalidos = [
    (300, 0, 0),     # componente > 255
    "colornulo",     # nombre inexistente en matplotlib
    "#ZZZZZZ",      # hex inv√°lido
    42,             # tipo entero ‚Äî no soportado
]

print("Demostraci√≥n de validaci√≥n de colores inv√°lidos:\n")
for caso in casos_invalidos:
    try:
        resultado = _color_a_rgb(caso)
        print(f"  ‚úì {repr(caso):25s}  ‚Üí  {resultado}")
    except ValueError as e:
        print(f"  ‚úó {repr(caso):25s}  ‚Üí  {e}")
````
